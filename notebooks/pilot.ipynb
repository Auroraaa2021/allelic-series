```python
import pandas as pd
import pylab as plt
```

```python
df = pd.read_csv("./../data/APOB.txt", sep='\t', header=None)
del df[2]
del df[6]
df[8] = df[8].astype(int)
df.columns = ['SNP', 'CHROM', 'POS', 'A1', 'A2', 'ANN', 'AC', 'N']
df
```

```python
plt.figure(1, figsize=(4,3))
_ = plt.hist(df['AC'], 50)
plt.yscale('log')
plt.xlabel('Allele count')
plt.ylabel('Number of variants')
plt.tight_layout()
```
